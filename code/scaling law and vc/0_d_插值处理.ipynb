```python
import math
def ployinterp_column(data,s,n,LEN):
    if n<LEN:
        if math.isnan(s[n-1]) or math.isnan(s[n+1]) or s[n-1]==0 or s[n+1]==0:
            y = s[n]
        else:
            y = (s[n-1] + s[n+1])/2  # 取前一位和后一位的均值
#     y = y[y.notnull()]  # 剔除空值
        return(y)
    else:
        return(s[n])

for i in range(data.shape[0]):  # 遍历每一列
    for j in range(1,27):
        if math.isnan(data.iloc[i:i+1,3:3+27].values[0][j]) or data.iloc[i:i+1,3:3+27].values[0][j] == 0:  # 如果为空即插值
            data.iloc[i:i+1,3+j:4+j] = ployinterp_column(data,data.iloc[i:i+1,3:3+27].values[0],j,26)
            print('amount',(i,j))
    for j in range(1,19):
        if math.isnan(data.iloc[i:i+1,3+27*5:22+27*5].values[0][j]) or data.iloc[i:i+1,3+27*5:22+27*5].values[0][j] == 0:  # POP 如果为空即插值
            data.iloc[i:i+1,3+27*5+j:4+27*5+j] = ployinterp_column(data,data.iloc[i:i+1,3+27*5:22+27*5].values[0],j,18)
            print('pop',(i,j))
        if math.isnan(data.iloc[i:i+1,22+27*5:41+27*5].values[0][j]) or data.iloc[i:i+1,22+27*5:41+27*5].values[0][j] == 0:  # GDP 如果为空即插值
            data.iloc[i:i+1,22+27*5+j:23+27*5+j] = ployinterp_column(data,data.iloc[i:i+1,22+27*5:41+27*5].values[0],j,18)
            print('GDP',(i,j))
```
